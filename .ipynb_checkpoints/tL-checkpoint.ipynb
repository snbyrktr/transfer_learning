```python
import os
import cv2
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
# Klasörlerin adları
imdir=['Data']
folders = ['glioma_tumor', 'meningioma_tumor', 'normal', 'pituitary_tumor']
```

```python
# data ve etiket listeleri
data = []
labels = []
image_names = []
```

```python
img_size=(128,128)
```

```python
# Her klasörde dolaşma
for folder_ in folders:
    fold_path = os.path.join('Data', folder_)  # 'Data' klasöründe olduğunu varsayalım
    lab_val = folders.index(folder_)
    
    # Klasördeki tüm fotoğrafları al
    for dosya in tqdm(os.listdir(fold_path), desc=f"Processing {folder_}"):
        dosya_yolu = os.path.join(fold_path, dosya)
        
        # Fotoğrafı yükleme
        img = cv2.imread(dosya_yolu)
        
        # Boyutlandırma ve renk kanallarını düzeltme
        img = cv2.resize(img, img_size)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        # Veri listesine ekleme
        data.append(img)
        
        # Etiket listesine ekleme
        labels.append(lab_val)
        # Görsel isimlerini ekleme
        image_names.append(dosya)
```

```python
# Veri ve etiket listelerini numpy dizilerine dönüştürme
data = np.array(data)
labels = np.array(labels)
image_names = np.array(image_names)

# Veri ve etiketlerin şeklini kontrol etme
print("Veri şekli:", data.shape)
print("Etiket şekli:", labels.shape)
print("Görsel isimleri şekli:", image_names.shape)
```

```python
## Görselleri gösterme
def show_images(images, labels, names,num=100):
    plt.figure(figsize=(20, 40))  # Boyutları daha büyük yaparak daha fazla görsel sığdırabiliriz
    for i in range(num):
        plt.subplot(int(np.ceil(num / 10)), 10, i + 1)  # 10 görsel/satır
        plt.imshow(images[i])
        plt.title(f" Name: {names[i]}")
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# Tüm görselleri, etiketlerini ve isimlerini gösterme
show_images(data, labels, image_names, num=100)
```
